#### Some plotting basics with the matplotlib library

We will develop some python3 code using the [matplotlib](https://matplotlib.org/stable/plot_types/index.html#) and [numpy](https://numpy.org/doc/stable/reference/routines.math.html) libraries that can help with data visualization problems. When you are entering tokens into a cell that is set to code mode, you can use the <b>tab</b> key after you type the library name for more help! For example if you type numpy.random. and then hit <b>tab</b> you will see a graphical list of the functions and constants inside the numpy.random library like [random](https://numpy.org/doc/stable/reference/random/index.html#) . Try it out and see what types of functions are here. Some examples are shown below.

```python
#make some fake data-------------------------------------------------------------
import numpy as np
x      = np.random.uniform(-100,100,100)  #ten real-value samples between -100 and +100
y      = np.random.uniform(0,10,100)      #ten real-value samples between 0 and 10
sizes  = np.random.uniform(10,200,len(x)) #size of data points (x,y)

colors = np.zeros((len(x),),dtype=int)    #solid color=0
for i in range(len(colors)):              #old fashioned for loop way
    colors[i] += 90                       #increase each color by 90

print(x[:10])                             #print first 10 values
print(y[:10])                             #print first ten values
```

```python
import matplotlib.pyplot as plt        #visualization library
plt.style.use('_mpl-gallery')          #setup default styles
plt.rcParams['figure.figsize'] = [6,6] #set the default plot size

# plotting code here---------------------------------
def scatter(x,y,sizes,colors):
    fig, ax = plt.subplots()
    ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)
    ax.set(xlim=(-100,100), xticks=np.arange(-100,100,10),
           ylim=(0,10), yticks=np.arange(0,10,1))
    plt.show()

#function call!
scatter(x,y,sizes,colors) #normal solid colors

rcolors = np.random.uniform(10,200,len(x)) #make colors random
scatter(x,y,sizes,rcolors)
```

#### [5 20 pts] Put it all together and play around with color and size
Now you can play around with another random distributions from numpy such as [numpy.random.geometric](https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.geometric.html). To learn how to use one of these and to get some background you can type numpy.random. then tab and see a list of distributions.
```python
#your code here
plt.style.use('dark_background') # Setting up a style
plt.rcParams['figure.figsize'] = [7,7] # Setting up the plot size

# Using the numpy.random.beta function
a = np.random.beta(60,40,1000) 
b = np.random.beta(40,60,1000)

sizes  = np.repeat(5,len(a)) # Size of the data points (same for all of the data)

# Setting up the colors
colors = np.zeros((len(a),),dtype=int)
for i in range(len(colors)):
    colors[i] += 80    

# Defining a function to do the plotting
def scatter(x,y,sizes,colors):
    fig, ax = plt.subplots()
    ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)
    ax.set(xlim=(0.4,0.8), xticks=np.arange(0.4,0.8,0.05),
           ylim=(0.2,0.6), yticks=np.arange(0.2,0.6,0.05))
    plt.show()

# Coloring the values within the 0.05 range of the center values of axises
for i in range(len(a)):
    if abs(0.60-a[i])<=0.05 and abs(0.40-b[i])<=0.05:
        colors[i] = 30

# Calling to function to show the plot
scatter(a,b,sizes,colors)
```
